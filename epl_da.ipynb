#### **Penalty Percentage of the Goals scored**

```python
goals = df.Goals.sum()
penalty_goals = df['Penalty_Goals'].sum()
goals_without_pk = goals - penalty_goals
print(goals_without_pk, penalty_goals)
```

```python
values = [goals_without_pk, penalty_goals]
labels = ['Non Panalty Goals', 'Penalty Goals']
fig = go.Figure(data=[go.Pie(labels=labels, values=values, title='Penalty/Non-Penalty Goals')])
fig.show()
```

##### It is clear  that 10% of goals were scored by peanlties, but How much more could have scored by penalties let's check how much penalties were missed

```python
pk_not_scored = df['Penalty_Attempted'].sum() - penalty_goals
values = [pk_not_scored, penalty_goals]
labels = ['Penalties Missed', 'Penalties Scored']
fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

##### So almost 18% penalties were missed probably because of the smart moves by Goalkeeper or choke finish by the penalty takers if these were score the total goal tally of the league would have crossed thousand mark.

#### **Goals and Assist**

```python
assists = df.Assists.sum()
values = [goals-assists, assists]
labels = ['Goals without Assist', 'Goals with Assist']
fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

##### So with the above observation, The Above chart shows that 31% did not needthat final decisive pass.

#### **Highest Goalscorer for each Club**

```python
def getGoalscorer(row):
    name = df.loc[row.Goals]['Name']
    return name

def getHowMany(row):
    goals = df.loc[row.Goals]['Goals']
    return goals

goal_perclub=df['Goals'].groupby(df['Club']).idxmax().to_frame()
goal_perclub['Name'] = goal_perclub.apply(getGoalscorer, axis=1)
goal_perclub['Goals'] = goal_perclub.apply(getHowMany, axis=1)

goal_perclub.set_index('Name', inplace=True)
goal_perclub.sort_values('Goals', inplace=True, ascending=False)


fig = px.bar(goal_perclub, x=goal_perclub.index, y=goal_perclub['Goals'], color='Goals')
fig.update_layout(title="Top scorer from each club", xaxis_title="Player", yaxis_title="Goals")
fig.show()
```

##### Surprised to see Jorginho as Chelsea's highest goalscorer with less than 10 goals and I am so sure almost all came from penalty kicks. But one thing to figure out here is that they need a goalscorer, one to give you Harry Kane, Salah kind of goals if they are to compete for the league next season. Gundogan for Manchester City seems fair enough given that Pep rotates his squad often. Harry Kane is the Tottenham and EPL highest goalscorer surpassing the 20-goal mark with Salah.

#### **Goal Involvement by Players**

```python
def getHowMany(row):
    contrib=row.Goals+row.Assists
    return contrib
df['Goal_Involv']= df.apply(getHowMany, axis=1)
most_creat=df.sort_values(by='Goal_Involv',ascending=False)[:10]

fig = px.bar(most_creat, x='Name', y=['Goals', 'Assists'], barmode='group', title='The Most Creative', hover_data=('Club', 'Position', 'Age'))
fig.show()
```

##### Harry Kane is the most creative player appparently with 23 Goals and 14 Assists his team-mate Son Heung-min is the 4th most creative after Bruno Fernandes and Mohamed Salah even after this great performance ottenham was not able to reach Top 4, they ended season at 7th place.

##### These were the most creative players among all but now let's see who were the most creative players out of attcking roles.

```python
def_inv=df[~df.Position.str.contains('MF|FW', regex=True)].sort_values(by='Goal_Involv', ascending=False)[:10]
fig = px.bar(def_inv, x='Name', y=['Goals', 'Assists'], barmode='group', title='The Most Creative(Not Attacking Role)', hover_data=('Club', 'Position', 'Age'))
fig.show()
```

##### No surprise there Trent was one of the best players in pl in terms of adding value to team, not only he defends but he also provided a good amount of assists and scored 2 goals, Cresswell happens to be the defender with the most assists with 8 assists and no goal to his name. Digne and Coufal alongside Cresswell went through the entire season without a goal while Lewis Dunk had all goals but assist.

#### **Players with the Most Successful Passes to Assist**

```python
assistsmore1=df.loc[(df.Assists>5)]
def getPassesPerAssist(row):
    ppa=row.Successful_passes/row.Assists
    return ppa
assistsmore1['PassesPerAssist']= assistsmore1.apply(getPassesPerAssist, axis=1)
ppa_data=assistsmore1.sort_values(by='PassesPerAssist')[:10]
fig = px.bar(ppa_data, x='Name', y='PassesPerAssist', color='PassesPerAssist', title='Successful Passes to Assist')
fig.show();
```

##### This shows how many successful passes a player takes which would lead to a goal. Kevin DeBruyne might be the assist king but it takes him more than 80 successful passes to register an assist unlike Jamie Vardy, who takes up to 40 passes. Timo Werner might not be good when it comes to scoring but he sure supports as he needs up to 80 successful passes to register an assist. Imagine if he had a striking partner alongside him at Chelsea.
